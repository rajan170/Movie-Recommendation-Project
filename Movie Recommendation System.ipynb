```python
import ast
'''Converts dictionary to lists of lists'''
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L    
```

```python
import ast
'''Converts dictionary to lists of lists'''
def convert2(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter<4:  #to select first 4 characters and names only
#             L.append(i['name']+' : '+i['character'])
              L.append(i['name']+' '+i['character'])
              counter+=1
        else:
            break
    return L    
```

```python
movies_df['cast'] = movies_df['cast'].apply(convert2)
movies_df['cast']
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']  == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies_df['crew'].apply(fetch_director)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
ps.stem("action adventure fantasy sciencefiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d samworthingtonjakesully zoesaldananeytiri sigourneyweaverdr.graceaugustine stephenlangcol.quaritch jamescameron in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization.")
```

```python
def stem(text):
    L=[]
    for i in text.split():
        L.append(ps.stem(i)) 
    return " ".join(L)     
```

```python
new_df['tags'].apply(stem)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```
